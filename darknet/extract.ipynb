# 1080 x 2400のゲーム画像から左上400 x 400、右上400 x 400の画像を切り取る

```python
import keras.utils.image_utils as image
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator 
from tensorflow.keras.utils import array_to_img
import os
```

```python
file = "../../game_1080_2400.jpg"

print(os.path.basename(file))
print(os.path.dirname(file))
org_img = image.load_img(file)
plt.imshow(org_img)
org_img = np.array(org_img)
print("org_img shape")
print(org_img.shape)
```

```python
#矩形はりつけ
left_upper = org_img[0:399,0:399]
left_upper_img = array_to_img(left_upper, scale = False)
plt.imshow(left_upper_img)
```

```python
ymax = org_img.shape[0]
xmax = org_img.shape[1]
right_upper = org_img[0:399,xmax-400:xmax]
right_upper = array_to_img(right_upper, scale = False)
plt.imshow(right_upper)
```
